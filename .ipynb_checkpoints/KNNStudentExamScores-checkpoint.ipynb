```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("student_exam_scores.csv")
data.head()
```

```python
basarili_ogrenciler = data[data.exam_score >= 40]
basarisiz_ogrenciler = data[data.exam_score < 40]
plt.scatter(basarili_ogrenciler.hours_studied,basarili_ogrenciler.exam_score,color='green',label='Basarili',alpha=0.4)
plt.scatter(basarisiz_ogrenciler.hours_studied,basarisiz_ogrenciler.exam_score,color='red',label='Basarisiz',alpha=0.4)
plt.xlabel("Calisma Saati")
plt.ylabel("sınav notu")
plt.legend()
plt.show()
```

```python
y = data.exam_score
y = (y >= 40).astype(int)  
x_ham_veri = data.drop(["exam_score","student_id"],axis=1)

x = (x_ham_veri - np.min(x_ham_veri))/(np.max(x_ham_veri) - np.min(x_ham_veri))
print("Normalization öncesi ham veriler:\n")
print(x_ham_veri.head())

#Sonra
print("\n\nNormalization sonrası yapay zekaya eğitim için vereceğimiz veriler:\n")
print(x.head())
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=1)

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
print("k=3 için Test verilerimizin doğrulama testi sonucu %", knn.score(x_test, y_test)*100)
```

```python
sayac = 1
for k in range(1,11):
    knn_yeni = KNeighborsClassifier(n_neighbors = k)
    knn_yeni.fit(x_train,y_train)
    print(sayac," ","Doğruluk Oranı: %", knn_yeni.score(x_test,y_test)*100)
    sayac += 1
```
